```python
from datasets import load_dataset
import pandas as pd
import random

dataset = load_dataset("squad_kor_v1")
train = pd.DataFrame(dataset['train'])

# 중복된 context가 4개 이상인 행만 골라서
train = train[train['context'].map(train['context'].value_counts()) > 3]

new_rows = []

# 같은 context를 group으로 묶는다.
grouped = train.groupby('context')

for context, group in grouped:
    # 이때 group에서 상위 5개의 행만 남긴다.
    group = group.iloc[:5]
    seen = set()
    choices = []
    for ans in group['answers']:
        choice = ans['text'][0]
        if choice not in seen: # choice에 중복된 선택지가 있으면 제거한다.
            seen.add(choice)
            choices.append(choice)
    
    # 그리고 group에서 1개의 행을 랜덤으로 골라서 답으로 설정하고
    selected_row = group.sample(n=1).iloc[0]
    question = selected_row['question']
    answer_text = selected_row['answers']['text'][0]

    # 해당 행의 index+1을 answer에 저장한다.
    try:
        answer_index = choices.index(answer_text) + 1
    except ValueError:
        continue
    
    # problems dictionary를 정의한다.
    problems = {
        'question': question,
        'choices': choices,
        'answer': answer_index
    }
    
    # ID는 korquad-기존id 형식으로 정한다.
    new_id = 'korquad-' + str(selected_row['id'])
    
    new_row = {
        'id': new_id,
        'paragraph': context,
        'problems': problems,
        'question_plus': ''
    }
    
    new_rows.append(new_row)

new_df = pd.DataFrame(new_rows)

# 잘 됐나?
print(new_df.head())

# 잘 됐으니 저장하자.
new_df.to_csv('korquad.csv', index=False)
```
